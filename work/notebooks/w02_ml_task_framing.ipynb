## My Lane as an ML Task

I selected the Search Opportunity Ranking lane.

Task Type: Ranking

The goal is to rank pages based on their potential search opportunity so that the highest opportunity pages can be prioritized first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The target is search opportunity.

Since opportunity cannot be observed directly, a proxy target can be created using historical performance signals such as clicks, impressions, rankings, or engagement metrics available in the dataset.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

A useful success metric is Precision@K.

The model is successful if the pages ranked at the top are genuinely high-opportunity pages.

Higher Precision@K means better recommendations.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule may only consider one signal such as clicks.

Machine learning can combine multiple signals at the same time and identify patterns that are difficult to capture with manual rules.

This may produce more useful rankings than a single handcrafted rule.
